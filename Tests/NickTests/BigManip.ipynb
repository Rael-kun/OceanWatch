```python
numEntry = 14751772
# One big file
zero_arr = np.zeros(shape=[numEntry, 7])
globIndex = 0

# Normalization
latmin = 54.5
latmax = 56
lonmin = 13
lonmax = 16
latWhole = latmax - latmin
lonWhole = lonmax - lonmin
maxSpeed = 50
maxCourse = 360
pathIndex = 0

for i in range(len(data["indicies"])):
    # Read each traj
    tmpData = dataset_utils.readTrajectory(filename, i, superFileName)
    trackLen = tmpData["track_length"][0]
    for j in range(trackLen):
        # mmsi and timestamp
        zero_arr[globIndex][0] = tmpData["mmsi"][j]
        zero_arr[globIndex][1] = tmpData["timestamp"][j]

        # normalized lat, lon, speed, course
        zero_arr[globIndex][2] = (tmpData["lat"][j] - latmin) / latWhole
        zero_arr[globIndex][3] = (tmpData["lon"][j] - lonmin) / lonWhole
        zero_arr[globIndex][4] = tmpData["speed"][j] / maxSpeed
        zero_arr[globIndex][5] = tmpData["course"][j] / maxCourse
        # pathini
        zero_arr[globIndex][6] = pathIndex
        # update variables
        globIndex += 1
    pathIndex += 1
    if(i % 1000) == 0:
        print(i)
```

```python
ultimateData = pd.DataFrame(zero_arr, columns=["mmsi", "timestamp", "alat", "alon", "aspeed", "aheading", "pathIni"])
```

```python
ultimateData.to_csv("danish_train.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import pickle
import datetime
import torch
import glob

from Config import config
from utils import utils, dataset_utils
```

```python
ultimateData = pd.read_csv(r"C:\Users\DSU\Desktop\AllCurrentClasses\AISweden\DanishData\danish_train.csv")
```

```python
ultimateData["Elim"] = 0
```

```python
pathRemove = []
samePath = False
maxTime = 3600
lastTime = 0
pathOn = -1

# Remove invalid points (Speed>50, Timestep>3600)
for i in range(len(ultimateData["mmsi"])):
    row = ultimateData.iloc[i]
    # check pathOn
    if(pathOn != row["pathIni"]):
        # Path switched
        lastTime = row["timestamp"]
        pathOn = row["pathIni"]
        samePath = False

    # Speed > 50
    if(row["aspeed"] > 1):
        ultimateData.iloc[i, ultimateData.columns.get_loc("Elim")] = 1
    # Timestep > 3600
    if((row["timestamp"] - lastTime) > maxTime and samePath == False):
        # Remove path
        pathRemove.append(row["pathIni"])
        samePath = True

    if (i % 100000 == 0):
        print(i)
    
```

```python
# Remove paths that are too short?
pathCon = pd.DataFrame(columns=["pathIni", "pathCount"])
pathOn = -1
pathSize = 0

for i in range(len(goodData["pathIni"])):
    row = goodData.iloc[i]

    if(pathOn != row["pathIni"]):
        # path Switched
        # enter into table the size
        pathCon.loc[len(pathCon)] = [pathOn, pathSize]
        # change variables
        pathOn = row["pathIni"]
        pathSize = 0
    pathSize += 1
    if(i % 10000 == 0):
        print(i)
```
